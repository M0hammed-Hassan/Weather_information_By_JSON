```python
#Required libraries in our project.
import json
import requests
import csv
import pandas as pd
```

```python
with open("weather_file.csv","w",newline="") as file: #We will make a csv file for weather. 
    fname=["Weather_Code","Weather_Description","Temperature(C)","Humidity(%)","Visibility(m)","Pressure(mb)","Wind_Speed(km/h)"]
    writer=csv.DictWriter(file,fieldnames=fname)
    writer.writeheader()
    for i in range(10): #For collecting data 10 times.
        url="http://api.weatherstack.com/current?access_key=cce43aa3a3de500cc32c5df0e67c9408&query=Cairo" #Web site link.
        response=requests.request("GET",url)
        json_response=json.loads(response.text)
        weather_code=json_response["current"]["weather_code"] #For getting the weather code.
        weather_description=json_response["current"]["weather_descriptions"][0]  #For getting weather description.
        temp=json_response["current"]["temperature"] #For getting weather temperature.
        humidity=json_response["current"]["humidity"] #For getting weather humidity.
        visibility=json_response["current"]["visibility"] #For getting weather visibility.
        pressure=json_response["current"]["pressure"] #For getting weather pressure.
        wind_speed=json_response["current"]["wind_speed"] #For getting wind speed.
        data={"Weather_Code":weather_code,"Weather_Description":weather_description,"Temperature(C)":temp,"Humidity(%)":humidity,"Visibility(m)":visibility,"Pressure(mb)":pressure,"Wind_Speed(km/h)":wind_speed}
        writer.writerow(data)
```
